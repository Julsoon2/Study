```python
import cv2

img = cv2.imread('yellow_dust.png')

target_img = img.copy() # 원본 이미지에서 카피한 객체 만들기

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # 흑백 변환하기
ret, otsu = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU) # 이진화하기

contours, hierarchy = cv2.findContours(otsu, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE) # 윤곽선 검출하기
#이미지, 윤곽선 찾는 모드(mode), 윤곽산 찾는 근사치 방법(method)

COLOR = (0, 200, 0) # 녹색
cv2.drawContours(target_img, contours, -1, COLOR, 1) # 윤곽선 그리기
# 대상 이미지, 윤곽선 객체, 인덱스(-1이면 전체), 색깔, 두께
# 인덱스는 윤곽선 찾은 것들이 리스트로 저장되는데, 모든 윤곽선을 다 그릴지(-1) 혹은 특정 윤곽선만 그릴지 선택 가능

cv2.imshow('img', img) # img 라는 제목의 창으로 img 객체 show
cv2.imshow('gray', gray)
cv2.imshow('otsu', otsu)
cv2.imshow('contour', target_img)

cv2.waitKey(0) # 키 입력까지 특정 시간동안 대기. ms 단위. 0이면 키 입력까지 무한대기
cv2.destroyAllWindows() # 화면에 나타난 윈도우 종료
```
