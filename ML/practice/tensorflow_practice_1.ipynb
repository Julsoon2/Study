```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df = pd.read_csv('gpascore.csv')
df
```

```python
# 데이터 전처리 - 빈 부분은 평균값을 넣거나 행을 삭제하기
# data.isnull().sum()  # 각 column에서 null인게 몇개 있는지 세어줌
df = df.dropna() # nan 혹은 빈 칸이 있는 row를 제거해줌.
# dat.fillna(100) # 빈칸을 원하는걸로 채워줌. 평균, 앞뒤값 등등 다양한 방법 있으니 검색 ㄱㄱ
```

```python
y = df['admit'].values # np.array로 반환
```

```python
X = []
for idx, rows in df.iterrows():
    X.append([rows['gre'], rows['gpa'], rows['rank']])
X = np.array(X)
X
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([  # 딥러닝모델 생성.
    tf.keras.layers.Dense(64, activation = 'tanh'), # 첫번째 레이어. 숫자는 node 갯수 -> 맘대로 정하면 됨. 잘 나올때까지. 보통 2의 승수로 많이들 함. activation은 활성함수.
    tf.keras.layers.Dense(128, activation = 'tanh'), # 두번째 레이어. 활성함수 종류는 sigmoid, tanh, relu, softmax, leakyRelu 등등 많음. 모델에 따라 적절히 선택.
    tf.keras.layers.Dense(1, activation = 'sigmoid') # 세번째 레이어 -> 마지막 레이어는 결과물 받고 싶은거에 따라 디자인. 확률 한개를 예측할때니까 1개. 출력결과도 0~1 사이의 확률이어야  하니까 sigmoid
])
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# compile에서 optimizer 등의 요소 넣어주기.
# optimizer는 w1 <- w1 - alpha * de/dw 역전파에서 alpha값 조정해주는애
# optimizer 예로는 adam, adagrad, adadelta, rmsprop, sgd 등 있음.
# 확률로 분류하는 문제는 loss function으로 binary_crossentropy를 씀. classification 할때 그 식임.
# metrics는 어떤 요소로 평가할건지 정하는건데 보통 accuracy 쓰면 됨.
```

```python
model.fit(X, y, epochs = 1000) 
# X는 들어갈 요소, y 는 정답
# epochs는 학습시키고 싶은 횟수.
# X 데이터에 들어갈 형식
# [    [첫번째 row], [두번째 row], [세번째 row] , ...  ]   요렇게 들어가야함. 한 행이 하나의 array로 들어가고, 전체 array가 하나의 array로 X에 들어감.
# 데이터는 numpy array 형태로 집어넣어야 함.
# y는 그냥 1d array 하나로 들어감. 가끔씩 (1, 1) 형태로 들어가기도 함.
# 걸린시간. LOSS는 적어져야 좋은거. ACCURACY는 모델의 정답률. 높아져야 좋음.
```

```python
# 학습시킨 모델로 예측해보기.
# fit에 들어가는거랑 똑같은 형태로 데이터 넣으면 됨.
predict = model.predict(    [[750, 3.70, 3], [400, 2.2, 1]]   ) # 첫번째 애는 성적 좋고, 두번째 애는 구린애로 해봄
print(predict) # 처음 애는 84.8%로 붙고, 두번째애는 5%로 붙음.
```
