```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, label = 'cos', linestyle = "--")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sin & cos')
plt.legend(loc = 'upper right')
plt.show()
```
